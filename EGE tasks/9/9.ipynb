```python
f = open('9(1).txt').readlines()

n=0

for s in f:
    a = sorted([int(x) for x in s.split()])
    if ((sum(a)-min(a)-max(a))/3)>=8:
        n+=1
print(n)
```

```python
f = open('9(4).txt').readlines()

n=0

for s in f:
    a = sorted([int(x) for x in s.split()])
    pov = [x for x in a if a.count(x)==3]
    nepov = [x for x in a if a.count(x)==1]
    c = [x for x in a if x%2==0]
    nec = [x for x in a if x%2!=0]
    u=0
    if len(set(pov))==1 and len(nepov)==3: u+=1
    if len(c)>len(nec): u+=1
    if a[-2]+a[-1]>2*(sum(a)-(a[-2]+a[-1])): u+=1
    if u>=2:
        n+=1
print(n) 
```
