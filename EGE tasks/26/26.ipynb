```python
f = open('26_1.txt').readlines()[1:]
a = [int(x) for x in f]
mn =set(a)

k = []

for b in range(len(a)):
    for c in range(b+1, len(a)):
        if a[b]%2 == a[c]%2 and (a[b]+a[c]) in mn:
            k.append(a[b] + a[c])
            
print(len(k), max(k))
```

```python
f=open('26_3.txt').readlines()[1:]

a = [[0]*10_000 for _ in range(10000)]

for s in f:
    y,x = [int(x) for x in s.split()]
    a[y-1][x-1] = 1
    
mk = 0
t = 0
r = 0
for x in a:
    t+=1
    k=0
    for i in range(len(x)):
        if x[i] == 1: 
            k+=1
            if k>mk:
                mk = k
                r = t
        elif x[i] == 0:
            k = 0

print(mk, r)
```
